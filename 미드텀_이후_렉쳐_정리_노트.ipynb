> Difference between List and Tuple
>
>
> List
> * 데이터 사이즈가 큼
> * mutable(값을 바꿀 수 있음)
> * 리스트끼리 더했을 때 같은 id를 가지게 업데이트 됨
>
> Tuple
>
> * 데이터 사이즈가 작음 
> * inmutable(값을 바꿀 수 없음)
> * 튜플끼리 더했을 때 새로운 id를 가진 튜플을 생성

````python
mytuple = (1, 2)
mylist = [1, 2]

a = tuple(range(1000))
b = list(range(1000))
print(a.__sizeof__()) # 몇개의 바이트를 소모하는 지
print(b.__sizeof__())

b = [1, 2]
b[0] = 3

a = (1, 2)
# a[0] = 3 # tuple is immutable

# 튜플과 튜플을 더하고, 리스트와 리스트를 더했을 때 같은 메모리 주소를 갖는 지
a = (1, 2)
b = [1, 2]
print( id(a), id(b))

a += (3,)
b += [3]
print( id(a), id(b))

c = {a : 1}
#b = (b : 1) # error
````

## Dictionary

````python
# 딕셔너리 선언 방법
colors = {
    "Sam" : "Blue",
    "Amy" : "Red",
    "Sarah": "Yellow"
}
print(colors)
print(id(colors))

# 딕셔너리에서 키로 해당 value 찾기
print(colors["Sarah"])

# 딕셔너리에 있는 모든 키 값과 모든 value 값 찾기
print(colors.keys())
print(colors.values())

for item in colors.keys():
    print( "{0} likes the color{1}".format( item, colors[item] ))

# 딕셔너리에 있는 키 값에 해당되는 Value 값을 바꿀 때
colors["Sarah"] = "Purple"

# 딕셔너리에 새로운 키:Vaule 를 넣을 때
colors.update({
    "Harry" : "Orange"
})
print(colors)

# 딕셔너리에 저장 된 데이터 개수
print(len(colors))

# 딕셔너리 초기화
colors.clear()
print(len(colors))
print(id(colors))
````
